- To detect them, there is a beautiful method which is called `.isnull()`. With this method, we can get a boolean series (True or False).

```python
# Determine the number of missing values
# Syntax: DataFrame.isnull().sum()
combined_toxic_threeYear.isnull().sum()
```

```python
# Let's create a function to determine the percentage of missing values
# Typically less than five percent missing values may not affect the results
# More than 5% can be dropped, replaced with existing data, or imputed using mean or median.
# Syntax: def missing(DataFrame):
#    print ('Percentage of missing values in the dataset:\n',
#           round((DataFtame.isnull().sum() * 100/ len(DataFrame)),2).sort_values(ascending=False))

def missing(combined_toxic_threeYear):
    print ('Percentage of missing values in the dataset:\n',
           round((combined_toxic_threeYear.isnull().sum() * 100/ len(combined_toxic_threeYear)),2).sort_values(ascending=False))
# Call the function and execute
# Syntax: missing(DataFrame)
missing(combined_toxic_threeYear)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create a range of x values for three different pollutants
x1 = np.arange(-0.5, 3, 0.001)
x2 = np.arange(-0.5, 1, 0.001)
x3 = np.arange(-0.2, 0.5, 0.01)
# Create the corresponding y values for normal distributions with different means and standard deviations
y1 = norm.pdf(x1, 1.549783, 0.188114)   # Pollutant 1
y2 = norm.pdf(x2, 0.126859, 0.030100) # Pollutant 2
y3 = norm.pdf(x3, 0.000902, 0.000735) # Pollutant 3
# Create a figure and a 1x3 grid of subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
# Plot the bell curves for the three pollutants
axes[0].plot(x1, y1, label='Pollutant 1')
axes[0].set_title('Formaldehyde Density Plot')
axes[0].set_xlabel('Concentration (ug/cubic meter)')
axes[0].set_ylabel('Probability Density')
axes[1].plot(x2, y2, label='Pollutant 2', color='green')
axes[1].set_title('Benzene Density Plot')
axes[1].set_xlabel('Concentration (ug/cubic meter)')
axes[2].plot(x3, y3, label='Pollutant 3', color='red')
axes[2].set_title('Cyanide Density Plot')
axes[2].set_xlabel('Concentration (ug/cubic meter)')
# Adjust layout to prevent overlap
plt.tight_layout()
# Show the plots
plt.show()
```

* the extent to which those variables increase or decrease in parallel; a negative correlation indicates the extent to which one variable

* correlation Correlation is a mutual relationship or connection between two or more things. It takes a value between (+1) and (-1)
